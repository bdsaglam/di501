```python
### answer ###

import scipy
from scipy import stats

def plot_histograms(df, bins='auto'):
    cols = df.columns
    f, axes = plt.subplots(1, len(cols), figsize=(8*len(cols), 8))
    axes = axes if len(cols) > 1 else [axes]
    for i, ax in enumerate(axes):
        x = df[cols[i]]
        hist, bin_edges = np.histogram(x, bins=bins)
        sns.histplot(x, bins=bins, ax=ax, kde=True)
        ax.set_title(f"Histogram with bins={bins}")
        ax.set_xlabel(f"{cols[i]}")

        # Plot the PDF.
        xmin, xmax = ax.get_xlim()  # extract the endpoints for the x-axis
        mu, std = stats.norm.fit(x)
        xn = np.linspace(xmin, xmax, 100)
        pn = stats.norm.pdf(xn, mu, std) * hist.sum()
        ax.plot(xn, pn, 'k', linewidth=2, label='normal')
        ax.legend()
    
    return f, axes

# From Sturge's formula
bins_recommended = int(np.ceil(1 + 3.322 * np.log10(len(popcorn_df))))
for bins in [bins_recommended, 5, 10, 20, 50, 100]:
    plot_histograms(popcorn_df[['popcorn', 'age']], bins=bins)
```

<h5 style="margin-bottom:0">e. Draw a heatmap of correlations (numerical variables) in our dataset. </h5>

Draw a heatmap of correlations by using seaborn. Set your DPI to 1200. Then, interpret the correlation coefficient between age and popcorn sales. Please save your image on your computer as a PNG file as well.
