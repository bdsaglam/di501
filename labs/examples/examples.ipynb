```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy.stats import pearsonr
import statsmodels.api as sm 
import pylab as py 
```

```python
df1=pd.read_csv("df1.csv")
df2=pd.read_csv("df2.csv")
df3=pd.read_csv("df3.csv")
df4=pd.read_csv("df4.csv")
df5=pd.read_csv("df5.csv")
df6=pd.read_csv("df6.csv")
df7=pd.read_csv("df7.csv")
df8=pd.read_csv("df8.csv")
```

```python
def visualize_df(df):
    rows=len(df.columns)
    f, axes = plt.subplots(rows, 4, figsize=(20,8))
    for i in range(2):
        sns.histplot(df.iloc[:,i], ax=axes[i,0])
        sns.boxplot(x=df.iloc[:,i], ax=axes[i,1])
        sns.violinplot(x=df.iloc[:,i], ax=axes[i,2])
        sm.qqplot(df.iloc[:,i], line="s", ax=axes[i,3])
    plt.show()  
```

```python
def analyze_corr(df):
    plt.title("\nScatterplot of x1 & x2\n", color="purple")
    plt.scatter(df["x1"],df["x2"])
    corr, p_value = pearsonr(df["x1"],df["x2"])
#     print("pearsons correlation: %.4f" % corr)
    if(corr>=0.50):
        print("pearsons correlation: %.4f" % corr,">> strong positive association")
    elif( (corr<0.50) & (corr>=0.30)): 
        print("pearsons correlation: %.4f" % corr,">> positive medium association")   
    elif( (corr<0.30) & (corr>=0.10)):
        print("pearsons correlation: %.4f" % corr,">> positive small association") 
    elif( (corr<0.10) & (corr>-0.10)):
        print("pearsons correlation: %.4f" % corr,">> no association")       
    elif( (corr<-0.10) & (corr>-0.30)):
        print("pearsons correlation: %.4f" % corr,">> negative small association")
    elif( (corr<=-0.30) & (corr>-0.50)):
        print("pearsons correlation: %.4f" % corr,">> negative medium association")
    elif( corr<=-0.50):
        print("pearsons correlation: %.4f" % corr,">> strong negative association")

```

```python
analyze_corr(df1)
visualize_df(df1)
```
